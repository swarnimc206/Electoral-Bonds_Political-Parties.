```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(data.head())

# Check basic information about the dataset
print("\nBasic information about the dataset:")
print(data.info())

# Convert 'Date of Encashment' column to datetime datatype
data['Date of\nEncashment'] = pd.to_datetime(data['Date of\nEncashment'])


```

```python
print(data['Denomination'].head(10))
data['Denomination'] = data['Denomination'].astype(str)

import re

def clean_denomination(denomination):
  cleaned_denomination = re.sub(r'[^\d\.]', '', denomination)
  return float(cleaned_denomination)

data['Denomination'] = data['Denomination'].apply(clean_denomination)

median_denomination = data['Denomination'].median()

data['Denomination'].fillna(median_denomination, inplace=True)

print("\nNull values after handling:")
print(data.isnull().sum())
```

```python
# Exploratory Data Analysis (EDA)
# Plot histogram for Denomination
plt.figure(figsize=(10, 6))
data['Denomination'] = data['Denomination'].astype('str')
sns.histplot(data['Denomination'], bins=20, kde=True, color='skyblue')
plt.title('Histogram of Denomination')
plt.xlabel('Denomination')
plt.ylabel('Frequency')
plt.show()

```

```python


# Increase the figure size
plt.figure(figsize=(15, 8))

# Plot the data
sns.lineplot(x='Date of\nEncashment', y=data.index, data=data, estimator=None, marker='o', color='black')

# Set the title and axis labels
plt.title('Encashments Over Time (Month-wise)')
plt.xlabel('Date of Encashment')
plt.ylabel('Number of Encashments')

# Rotate the x-axis tick labels
plt.xticks(rotation=45)

# Tighten the layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
# Plot Encashments by Political Party
plt.figure(figsize=(12, 8))
sns.countplot(y='Name of the Political Party', data=data, order=data['Name of the Political Party'].value_counts().index, palette='viridis')
plt.title('Encashments by Political Party')
plt.xlabel('Number of Encashments')
plt.ylabel('Political Party')
plt.tight_layout()
plt.show()
```
